#### Capstone Project EDA

In this script, we perform basic EDA on each of the provided datasets to determine their size, features, and potential relevance for future analysis

Datasets include: appearances, club games, clubs, competitions, games, game events, game linueps, player valuations, players, and transfers

```python
import pandas as pd

#Load and examine transfer data set
transfers = pd.read_csv("data/transfers.csv")
transfers.head()

#Dimensions of the data
transfers.shape

#Look at missing values
transfers.isna().sum()

#Summary stats for numeric columns
transfers.describe()

#Look at observation with the highest transfer fee
transfers[transfers['transfer_fee'] == 180000000.0]

#Look at observation with the highest market value
transfers[transfers['market_value_in_eur'] == 180000000.0]
```
